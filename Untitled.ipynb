```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
sns.set_theme(style = "whitegrid")
```

```python
flare1 = datetime.datetime(2017, 9, 6, 8, 57)
flare2 = datetime.datetime(2017, 9, 6, 11, 53)
```

```python
df = pd.read_csv("./sample_features/segmented_7115.csv")
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values(by = "date")
print(df.columns)

key = "nl_rho_mean"

#df[key] = (df[key] - df[key].min()) / (df[key].max() - df[key].min())

ax = sns.lineplot(data = df, x = "date", y = key, palette="tab10")
ax.axvline(flare1, 0, 1, color = "red")
plt.xticks(rotation = 30)
plt.show()
```

```python
df = pd.read_csv("./sample_features/baseline_7115.csv")
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values(by = "date")

key = "totrho"

#df[key] = (df[key] - df[key].min()) / (df[key].max() - df[key].min())

df["totrho"] = df["totrho"] / (8 * np.pi)

ax = sns.lineplot(data = df, x = "date", y = key, palette="tab10")
ax.axvline(flare1, 0, 1, color = "red")
plt.xticks(rotation = 30)
plt.show()
```
